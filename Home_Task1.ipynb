```python
#Data Analytics: [0.3 Marks]
import pandas as pd # here we are importing libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('train.csv')
df.head()
```

```python

sns.countplot(x='Survived', data=df)
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()
```

```python
# Group data by Passenger Class and Gender, then sum the number of survivors
grouped_data = df.groupby(['Pclass','Sex'])['Survived'].sum()

# Convert grouped data into table format (Gender becomes columns)
sorted_list = grouped_data.unstack()

# Plot bar chart
sorted_list.plot(kind='bar', color=['blue','green'])

plt.title('Count of Survival by Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Survivors')
plt.xticks(rotation=0)
plt.show()
```

# QUESTION : Can we remove outliers or apply normalization/standardization on complete table?
 No, we cannot apply normalization or remove outliers on the whole table.
 This is because the dataset has both numbers and text data.
 These methods only work on numbers like Age or Fare.
 They do not work on text columns like Name or Sex.
 So, we should only use these methods on numerical columns, not the entire dataset.

```python
# Q.3 How do we change column names in Dataframe?
# Ans We can rename columns using the rename() function.

df.rename(columns={'Pclass':'Passenger_Class'}, inplace=True)
df.head()
```

```python
# Q.4 Apply logical conditions to print values greater than certain threshold
# Ans We can filter data using conditions.

df[df['Age'] > 30]
```

```python
# Q.5 What is the sort function in pandas?
# Ans Sorting is used to arrange data in order (ascending or descending).

# Ascending order
df.sort_values(by='Age', ascending=True)

# Descending order
df.sort_values(by='Age', ascending=False)
```

```python
# Q.6 How do I access data in a dataframe? (loc and iloc)
# Ans loc: Access using column name (label-based).
#     iloc: Access using index position (number-based).

# Using loc
df.loc[0, 'Age']

# Using iloc
df.iloc[0, 3]
```
